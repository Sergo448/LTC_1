```python
# LTC\main.py

# importing libraries
import pandas as pd

# path to xlsx file
file_path: str = 'P:\Python Projects\LTC_1\SN-2012\Глава 1. Здания.xlsx'

    
# reading file
col_names = ['Col1', 'Col2', 'Col3', 'Col4', 'Col5', 'Col6', 'Col7', 'Col8', 'Col9', 'Col10', 'Col11', 'Col12', 
             'Col13', 'Col14', 'Col15', 'Col16', 'Col17', 'Col18', 'Col19', 'Col20', 'Col21']
file_xlsx = pd.read_excel(file_path, names=col_names)
```

```python
data = file_xlsx.copy()
```

```python
def appender(arr):
    # выделение нужного столбца
    # удаление прочей информации
    # выжимка номеров нормативов и кодов ресурсов
    orders_aray = []
    for order in arr[arr['Col3'].notna()]['Col3'].reset_index(drop=True):
        if len(order) > 1:
            orders_aray.append(order)
    orders_aray.pop(0)
    return orders_aray

appender(data)
```
